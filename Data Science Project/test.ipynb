```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Create the dataframe
data = {
    'Group': ['Kids', 'Kids', 'Kids', 'Kids', 'Kids', 'Kids', 'Kids', 
              'Men', 'Men', 'Men', 'Men', 'Men', 'Men', 'Men', 
              'Seniors', 'Seniors', 'Seniors', 'Seniors', 'Seniors', 'Seniors', 'Seniors', 
              'Women', 'Women', 'Women', 'Women', 'Women', 'Women', 'Women'],
    'State': ['NSW', 'NT', 'QLD', 'SA', 'TAS', 'VIC', 'WA',
              'NSW', 'NT', 'QLD', 'SA', 'TAS', 'VIC', 'WA',
              'NSW', 'NT', 'QLD', 'SA', 'TAS', 'VIC', 'WA',
              'NSW', 'NT', 'QLD', 'SA', 'TAS', 'VIC', 'WA'],
    'Sales': [18587500, 5700000, 8510000, 14515000, 5775000, 26360000, 5625000,
              19022500, 5762500, 8392500, 14655000, 5757500, 26407500, 5752500,
              18187500, 5465000, 8190000, 14717500, 5650000, 26315000, 5512500,
              19172500, 5652500, 8325000, 14970000, 5577500, 26482500, 5262500]
}

df = pd.DataFrame(data)

# Create the main plot with Seaborn
sns.set(style="whitegrid")
fig, ax1 = plt.subplots(figsize=(14, 7))

# Create a bar plot
sns.barplot(data=df, x='State', y='Sales', hue='Group', ax=ax1)
ax1.set_xlabel('State')

# Create a secondary x-axis
ax2 = ax1.twiny()
ax2.set_xlim(ax1.get_xlim())
# Assuming we want to show the unique groups on the secondary axis
group_labels = df['Group'].unique()
group_positions = [df[df['Group'] == group]['State'].index[0] for group in group_labels]

ax2.set_xticks(group_positions)
ax2.set_xticklabels(group_labels)
ax2.set_xlabel('Group')

plt.show()
```
